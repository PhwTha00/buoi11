# Visualizing Company Sale data

### Step 1. Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter

# set this so the graphs open internally
%matplotlib inline
```

### Step 2. Import the dataset from this [address](https://github.com/thieu1995/csv-files/blob/main/data/matplotlib/company_sales_data.csv).

```python
url = "https://raw.githubusercontent.com/thieu1995/csv-files/main/data/matplotlib/company_sales_data.csv"
data = pd.read_csv(url)
print(data)
```

### Exercise 1: Read Total profit of all months and show it using a line plot

![img](https://github.com/thieu1995/pandas/blob/main/11_matplotlib/company-sale/img/01.png?raw=1)

```python
company_profit_per_month = data['total_profit']
plt.plot(data['month_number'], company_profit_per_month, color='blue')
plt.title('Company Profit per Month')
plt.xlabel('Month')
plt.ylabel('Profit')
plt.xticks(data['month_number'])
plt.show()
```

### Exercise 2: Get total profit of all months and show line plot with the following Style properties

![img](https://github.com/thieu1995/pandas/blob/main/11_matplotlib/company-sale/img/02.png?raw=1)

```python
plt.plot(data['month_number'], data['total_profit'], linestyle='dashed', linewidth=3, marker='o', markersize=8, color='r', markerfacecolor='black', markeredgewidth=2, markeredgecolor='r')
plt.title('Company Sales data of last year')
plt.xlabel('Month')
plt.ylabel('Total Profit')
plt.xticks(data['month_number'])
plt.show()
```

### Exercise 3: Read all product sales data and show it  using a multiline plot

![img](https://github.com/thieu1995/pandas/blob/main/11_matplotlib/company-sale/img/03.png?raw=1)

```python
plt.plot(data['month_number'], data['facecream'], label='Face Cream', marker='o')
plt.plot(data['month_number'], data['facewash'], label='Face Wash', marker='o')
plt.plot(data['month_number'], data['toothpaste'], label='Toothpaste', marker='o')
plt.plot(data['month_number'], data['bathingsoap'], label='Bathing Soap', marker='o')
plt.plot(data['month_number'], data['shampoo'], label='Shampoo', marker='o')
plt.plot(data['month_number'], data['moisturizer'], label='Moisturizer', marker='o')

plt.title('Sales Data')
plt.xlabel('Month')
plt.ylabel('Sales Units')
plt.legend()
plt.xticks(data['month_number'])
plt.show()
```

### Exercise 4: Read toothpaste sales data of each month and show it using a scatter plot
![img](https://github.com/thieu1995/pandas/blob/main/11_matplotlib/company-sale/img/04.png?raw=1)

```python
plt.scatter(data['month_number'], data['toothpaste'], color='b', marker='o')
plt.title('Toothpaste Sales Data')
plt.xlabel('Month')
plt.ylabel('Sales Units')
plt.grid(linestyle='dashed')
plt.xticks(data['month_number'])
plt.show()
```

### Exercise 5: Read face cream and facewash product sales data and show it using the bar chart

![img](https://github.com/thieu1995/pandas/blob/main/11_matplotlib/company-sale/img/05.png?raw=1)

```python
plt.bar(data['month_number'], data['facecream'], label='Face Cream sales data', color='b', width=0.4)
plt.bar(data['month_number'] + 0.4, data['facewash'], label='Face Wash sales data', color='orange', width=0.4)

plt.title('Face Cream and Face Wash Sales Data')
plt.xlabel('Month')
plt.ylabel('Sales Units')
plt.grid(linestyle='dashed')
plt.legend()
plt.xticks(data['month_number'])
plt.show()
```

### Exercise 6: Read sales data of bathing soap of all months and show it using a bar chart. Save this plot to your hard disk

![img](https://github.com/thieu1995/pandas/blob/main/11_matplotlib/company-sale/img/06.png?raw=1)

```python
plt.bar(data['month_number'], data['bathingsoap'], color='b', width=0.6)

plt.title('Bathing Soap Sales Data')
plt.xlabel('Month')
plt.ylabel('Sales Units')
plt.grid(linestyle='dashed')
plt.xticks(data['month_number'])
plt.savefig('bathing_soap_sales.png')
plt.show()
```

### Exercise 7: Read the total profit of each month and show it using the histogram to see the most common profit ranges

![img](https://github.com/thieu1995/pandas/blob/main/11_matplotlib/company-sale/img/07.png?raw=1)

```python
plt.hist(data['total_profit'], bins=10, color='blue')
plt.title('profit data')
plt.xlabel('Profit Range in dollar')
plt.ylabel('Frequency')

plt.show()
```

### Exercise 8: Calculate total sale data for last year for each product and show it using a Pie chart

![img](https://github.com/thieu1995/pandas/blob/main/11_matplotlib/company-sale/img/08.png?raw=1)

```python
total_sale_last_year = data.iloc[:, 1:-2].sum()

plt.figure(figsize=(8, 8))
plt.pie(total_sale_last_year, labels=total_sale_last_year.index, autopct='%1.1f%%', startangle=140)
plt.title('Sales Data')
plt.axis('equal')
plt.show()
```

### Exercise 9: Read Bathing soap facewash of all months and display it using the Subplot

![img](https://github.com/thieu1995/pandas/blob/main/11_matplotlib/company-sale/img/09.png?raw=1)

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(data['month_number'], data['bathingsoap'], marker='o', color='black')
plt.title('Sales data of a Bathingsoap')
plt.xlabel('Month Number')
plt.ylabel('Sales Units in number')


plt.subplot(1, 2, 2)
plt.plot(data['month_number'], data['facewash'], marker='o', color='r')
plt.title('Sales data of a facewash')
plt.xlabel('Month Number')
plt.ylabel('Sales Units in number')


plt.tight_layout()
plt.show()
```

### Exercise 10: Read all product sales data and show it using the stack plot

![img](https://github.com/thieu1995/pandas/blob/main/11_matplotlib/company-sale/img/10.png?raw=1)

```python
custom_colors = ['purple', 'blue', 'red', 'black', 'green', 'yellow']
plt.figure(figsize=(10, 6))
plt.stackplot(data['month_number'], data.iloc[:, 1:-2].T, labels=data.columns[1:-2], colors=custom_colors)
plt.title('All Product Sales Data using stack plot')
plt.xlabel('Month')
plt.ylabel('Sales Units')
plt.legend(loc='upper left')
plt.xticks(data['month_number'])
plt.tight_layout()
plt.show()
```
